## Introduction

atomman.unitconvert.unit is a dictionary that relates the string names of units to their scaling values. While you can access this directly, it is usually better to use one of [unitconvert's](atomman.unitconvert.ipynb) functions.

The underlying code can be found in [atomman/unitconvert.py](../../atomman/unitconvert.py).

```python
import atomman.unitconvert as uc
```

Default length values are in Angstroms

```python
print 'angstrom -> ', uc.unit['angstrom']
print 'm        -> ', uc.unit['m']
```

List all unit names recognized by unitconvert

```python
for key in uc.unit:
    print key
```
