## Introduction

The interpolate_contour function creates a continuous 2D color contourmap for a specified atomic property over a defined range by interpolating values between the atomic positions. Primarily, this function is defined for use with values computed by atomman.defect.nye_tensor, but it can be used with any per-atom property to view the property as a continuous distribution.

__Notes__: the current version is incomplete. property_unit does not work, and the plotting is limited to z-projections.

Arguments:

- __system__ -- the atomman.System containing the atoms with the per-atom property that you want to plot.

- __property__ -- the name of the per-atom property to plot. 

Keyword Arguments:

- __index__ -- the index or indices of vector/tensor properties to plot. Default value is None (property is a per-atom scalar).

- __plotbounds__ -- a 3x2 numpy array specifying the Cartesian coordinate range to include in the plot. Default value is None (plot all atoms).

- __bins__ -- specifies how many grid points to interpolate by. The resulting grid will have bins x bins dimensions. Default value is 200.

- __dots__ -- Boolean flag, which if True will overlay the colormap with dots corresponding to the atomic positions. Default value is True.

- __czero__ -- Boolean flag affecting the color range of the plotted values. If True, the max and min color values are given opposite signs and the same absolute values such that value = 0 corresponds to the mid value. If False, the max and min color values are based solely on the range of values. Default value is True.

- __save__ -- Boolean flag, which if True will save the plot as property[index].png. Default value is False.

- __show__ -- Boolean flag, which if True will show the plot. Default value is True.

- __length_unit__ -- defines the unit of length to use in plotting the atomic coordinates. Default value is 'angstrom'.

- __property_unit__ -- defines the unit to use for the plotted property value. Default value is None (no conversion, units).

- __cmap__ -- the matplotlib colormap style to use for the property values colors. Default value is 'jet'.


The function also returns two values, the integrated area summation, and the integrated area mean of the property over the range plotted.

The underlying code can be found in [atomman/plot/interpolate_contour.py](../../atomman/plot/interpolate_contour.py).

## Demonstration

Library imports

```python
#Standard libraries
from __future__ import print_function

#http://www.numpy.org/
import numpy as np             

#http://matplotlib.org/
%matplotlib inline

#https://github.com/usnistgov/atomman
import atomman as am 
```

Load system containing an fcc screw dislocation

```python
system_dump_file = 'files/atomman.plot.interpolate_contour/Al-fcc-screw.dump'
system = am.load('atom_dump', system_dump_file)[0]
alat = 4.04527
```

List per-atom properties

```python
print(system.atoms_prop())
```

Plot potential energy

```python
plot_range = np.array([[-17, 12], [-15, 15], [0, alat*6**0.5]])
```

```python
am.plot.interpolate_contour(system, 'c_peatom', plotbounds=plot_range)
```

This plot is not meaningful with czero normalization, so set czero=False

```python
am.plot.interpolate_contour(system, 'c_peatom', plotbounds=plot_range, czero=False)
```

Now, let's try a different colormap

```python
am.plot.interpolate_contour(system, 'c_peatom', plotbounds=plot_range, czero=False, cmap='Reds')
```

We can also remove the atom dots

```python
am.plot.interpolate_contour(system, 'c_peatom', plotbounds=plot_range, czero=False, cmap='Reds', dots=False)
```
