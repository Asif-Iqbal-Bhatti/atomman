## Introduction

The atomman.defect.Stroh class solves the Eshelby anisotropic elasticity solution for a perfectly straight dislocation or crack using the Stroh method. This allows for the evaluation of a dislocation's elastic energy, as well as the ability to generate a dislocation monopole system. 

The underlying code can be found in [atomman/defect/Stroh.py](https://github.com/usnistgov/atomman/blob/master/atomman/defect/Stroh.py).

#### Library Imports

```python
%matplotlib inline
import os
import atomman as am
import numpy as np

import matplotlib.pyplot as plt
```

```python
directory = 'E:/calculations/dislocation-vacancy/dislocation-vacancy/bcc{110}[111]/Mo_MEAM'
a = 3.1674
save_file = '2012--Park-H--Mo.png'
#save_file = None
burgers = a * np.array([0.0, 0.0, 3.**(0.5) / 2.])
```

```python
base_system, symbols = am.load('atom_data', os.path.join(directory, 'base.dat'))
disl_system, symbols = am.load('atom_dump', os.path.join(directory, 'disl.dump'))
neighbor_list_cutoff = 0.95 * a
base_system.nlist(neighbor_list_cutoff)
```

```python
plot_range = np.array([[-10, 10],
                       [-10, 10],
                       [0, a * 3.**(0.5) / 2.]])
plot_scale = 1.885618083

am.defect.differential_displacement(base_system, disl_system, burgers, plot_range, 
                                    plot_scale=plot_scale,
                                    save_file=save_file)
```
